### Desafio 8

Escreva um script Python para encontrar as 10 palavras mais longas em um arquivo.

O arquivo TXT está localizado na mesma pasta do projeto (texto.txt).

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
palavras_comprimento = {}

with open('/content/drive/My Drive/Awari/Exercicios/br_python_challenges/texto.txt') as f:
  for linhas in f:
    palavras = map(lambda p: p.strip(',)(?'), linhas.replace('.', ' ').split())
    for palavra in palavras:
      palavras_comprimento.setdefault(palavra, len(palavra))
```

```python
i_top = 0
for palavra, comprimento in sorted(palavras_comprimento.items(), key = lambda x: -x[1]):
  if i_top <= 9:
    print(palavra, comprimento)
  i_top += 1
```

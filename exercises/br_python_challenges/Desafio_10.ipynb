### Desafio 10

Dada uma lista, divida-a em 3 partes iguais e reverta cada lista.

Exemplo: 

sampleList = [11, 45, 8, 23, 14, 12, 78, 45, 89]

Parte 1 [8, 45, 11]

Parte 2 [12, 14, 23]

Parte 3 [89, 45, 78]

```python
# Seu código

sampleList = [11, 45, 8, 23, 14, 12, 78, 45, 89]
```

```python
def dividir_lista(lista):
  chunks = []
  for i in range(0, len(lista), 3):
    yield lista[i:i + 3][::-1]
```

```python
for i, chunks in enumerate(dividir_lista(sampleList), start = 1):
  print(f'Parte {i}', chunks)
```
